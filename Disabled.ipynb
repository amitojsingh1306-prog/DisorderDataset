```python
import pandas as pd

df = pd.read_csv('Mental Health Dataset.csv')
df.head()
```

```python
df.head()
df.info()
df.describe()
df.isnull().sum()
```

## Task 1: Data Cleaning

```python
df[['Gender','Country','Occupation']].isnull().sum()
```

## Replace missing categorical values with mode

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Country'] = df['Country'].fillna(df['Country'].mode()[0])
df['Occupation'] = df['Occupation'].fillna(df['Occupation'].mode()[0])
```

## Validate Cleaning

```python
df[['Gender','Country','Occupation']].isnull().sum()
```

## Task 2: Data Transformation

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Gender_encoded'] = le.fit_transform(df['Gender'])
df['Country_encoded'] = le.fit_transform(df['Country'])
df['Occupation_encoded'] = le.fit_transform(df['Occupation'])
```

## Encode Disorder_Type and Gender using LabelEncoder.


```python
df.head()
```

## Task 3: String Manipulation

## Clean disorder names with .str.strip() and .str.title().

```python
df['Country'] = df['Country'].str.strip().str.title()
df['Occupation'] = df['Occupation'].str.strip().str.title()
```

## Identify disorders starting with “D” using regex '^D'

```python
countries_D = df[df['Country'].str.contains('^D', regex=True)]
countries_D.head()
```

## Count unique disorder types per region.
## �� Tip: Export summary to CSV.

```python
occupation_count = df.groupby('Country')['Occupation'].nunique()
occupation_count
```

## Task 4: Data Handling

## Group by Region/country and calculate disorder frequency.

```python
country_counts = df.groupby('Country').size()
country_counts
```

## Find most common disorder type.

```python
df.rename(columns={'Mental_Health_History':'Disorder_Type'}, inplace=True)
```

```python
df['Disorder_Type'] = ((df['Growing_Stress'] == 'Yes') |
                       (df['Mood_Swings'] == 'Yes') |
                       (df['Coping_Struggles'] == 'Yes')).map({True:'Yes', False:'No'})
```

```python
df['Disorder_Type'].value_counts()
```

```python
country_disorder = df.groupby(['Country','Disorder_Type']).size().reset_index(name='Count')
```

```python
country_disorder.head()
```

## bar plot using seaborn library

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data=country_disorder,
            x='Country',
            y='Count',
            hue='Disorder_Type')

plt.xticks(rotation=90)
plt.title("Disorder Frequency by Country")
plt.show()
```
